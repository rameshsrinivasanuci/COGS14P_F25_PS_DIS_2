1. Write a function **play_scale** that makes and plays the **first five notes** composing one octave in either ascending or descending order. 

I have imported all the functions in sound.py so you can call them.  If you are not able to, its because sound.py is not in the same folder as this assignment.  

The function `pop_scale()` in the module `sound` returns a dictionary with all the frequencies. You can get a list of all the frequencies in the first octave using the keyword 'frequencies' 

The first note A has a frequency of 440 in the first octave. In the second octave at 880, and all the other frequencies also double. The third octave starts at double that frequency or 1760 (4*440), and all the other frequencies are multiplied by 4.  So for any note we can get the equivalent frequency at any octave using

    frequency = frequency*2**(n_octave-1)


The function **play_scale** should have four input arguments, one is the octave (**n_octave**), the second is the **duration** of each note, and the third is an optional **volume** which defaults to 0.05, and the fourth is an optional order which is a string which is either 'ascend' or 'descend', and defaults to 'ascend'. If an invalid order is given it should not play anything and should print out "Invalid Order"
 

```python
# YOUR CODE HERE
print(pop_scales())
def play_scale(n_octave, duration, volume=0.05, order='ascend'):
    scales = pop_scales() #we set the dictionary to scales 
    freqs = scales['frequencies'] #we set the variable freq equal to the index defined in the dicitonary by 'frequencies'
    first_five_notes = freqs[:5] #here we call the first five notes of the frequency, or the first 5 indexs
    first_five_notes = first_five_notes * (2 ** (n_octave - 1)) #take the octave and subtract by 1 then multiply by 2 cause thats what an octave is, we then multiply by our frequencies to get the correct octave 
    if order == 'ascend':
        ordered = first_five_notes #if order is ascend then order is the first five notes 
    elif order == 'descend':
        ordered = first_five_notes[::-1] #if the order is descend the order is the first five notes backwards
    else:
        print("Invalid Order") #if the order is not either of those print invalid order 
        return
    for f in ordered: 
        tone = make_tone(f, duration) #for what we set the order of our first five notes to, we set the function make_tone for the duration and the notes
        play_sound(tone, volume) #then we use the play_sound function, with our arguments tone and volume to get the actual sound 
```

```python
#should play ascending order 2nd octave
play_scale(2,0.5)
```

```python
#should play descending order 1st octave
play_scale(1,0.5,order='descend')
```

```python
#should play descending order louder volume 
play_scale(2,0.5,volume = 0.15,order='descend')
```

2. Make a function **random_scales** which randomly plays an ascending or descending scale in a random octave chosen from the first, second or 3rd octave.  

- You should make use of **play_scale** that you created in problem 1.  

- **random_scale** should get as input arguments the **duration**, and an optional parameters for volume which defaults to 0.05.

- It should choose **n_octave** randomly from 1,2,3.  

- The function should play the scale and return a string with the type of scale of played, 'ascend' or 'descend'

- The function should also return which octave was played

Hint: use the random number generator `rng` to randomly select the type of scale and octave to play

```python
# YOUR CODE HERE
def random_scale(duration, volume=0.05):
    n_octave = rng.choice([1, 2, 3]) #randomly choose the octave from one of these indexes
    order = rng.choice(['ascend', 'descend']) #randomly choose if we ascend or descend
    play_scale(n_octave, duration, volume, order) #play_scale function, random octave, set duration, set volume, random order 
    return order, n_octave
```
