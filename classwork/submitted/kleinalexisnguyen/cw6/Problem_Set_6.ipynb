1. Write a function **play_scale** that makes and plays the **first five notes** composing one octave in either ascending or descending order. 

I have imported all the functions in sound.py so you can call them.  If you are not able to, its because sound.py is not in the same folder as this assignment.  

The function `pop_scale()` in the module `sound` returns a dictionary with all the frequencies. You can get a list of all the frequencies in the first octave using the keyword 'frequencies' 

The first note A has a frequency of 440 in the first octave. In the second octave at 880, and all the other frequencies also double. The third octave starts at double that frequency or 1760 (4*440), and all the other frequencies are multiplied by 4.  So for any note we can get the equivalent frequency at any octave using

    frequency = frequency*2**(n_octave-1)


The function **play_scale** should have four input arguments, one is the octave (**n_octave**), the second is the **duration** of each note, and the third is an optional **volume** which defaults to 0.05, and the fourth is an optional order which is a string which is either 'ascend' or 'descend', and defaults to 'ascend'. If an invalid order is given it should not play anything and should print out "Invalid Order"
 

```python
# YOUR CODE HERE stop forgetting to define things
def play_scale(n_octave, duration, volume=0.05, order='ascend'):
    notes = pop_scales() # dictionary 1st octive (all the ntoes!)
    first_oct = notes['frequencies'][0:5] #1st 5 notes from 1st octave

    frequency = 2**(n_octave - 1) * first_oct # first octave from popscales() * 2^(whatever given octave)

    if order == 'ascend': #default
        final_freqs = frequency
    elif order == 'descend':
        final_freqs = frequency[::-1] #reverse order (like on exam2)
    else:
        print("Invalid order")
        return # have to put this here to stop the function if there's an invalid order... Can't use 'break' b/c not a loop

    for f in final_freqs:
        tone = make_tone(f, duration) # maketone() returns pure tone (sine wave of whatever frequency is being played)
        play_sound(tone, volume=volume) # play_sound() applies volume of whatever current tone at given volume.
```

3. Copy and paste your function from Problem Set 5, **get_response** that queries the subject with a prompt and checks if the response is among the correct keys. They should be prompted to respond again if they type any other key. You may use my function from the answer key.   

```python
# YOUR CODE HERE
def get_response(prompt, validresponses):
    response = input(prompt)
    is_valid = False
    while is_valid ==False:
        for valid in validresponses:
            if response ==valid:
                is_valid = True
                break
        if is_valid ==False:
            print("Invalid response, try again")
            response = input(prompt)
    print(response)
    return response
```

4. Write a function called **trial** that uses **random_scale** and **get_response** to play a random scale, then ask for a response of a for ascending and d for descending.

The subject should response with they key `a` for ascending and `d` for descending.  A good prompt might be

        'Respond a for ascending and d for descending: '

The function should evaluate if the response is correct or incorrect and create a variable **accuracy** which is 1 if they get it right and 0 if they get it wrong.

The function should return **accuracy**, **response**, and **order**

The input arguments for trial are **duration** and **volume**.  **volume** should have a default value of 0.05.   

```python
# YOUR CODE HERE
def trial(duration, volume=0.05):
    order, n_octave = random_scale(duration, volume) # random scale!!!!!!

    prompt = "Respond 'a' for ascending and 'd' for descending: "#defining prompt for get_response
    validresponses = ['a', 'd'] #defining valid responses for get_response
    response = get_response(prompt, validresponses) #call get response to get yer responses

    if order == 'ascend' and response == 'a': #if the order chosen by random_scale is ascend & participant inputs 'a', get accuracy point
        accuracy = 1
    elif order == 'descend' and response == 'd': #same goes for descend
        accuracy = 1
    else:
        accuracy = 0 #otherwise they get 0

    print('accuracy:', accuracy, ' response:', response, ' order:', order) #print stuff
    return accuracy, response, order #return statement
```
